–ó–¥–∞–≤–∞–ª–æ—Å—è –±, –≤–∏–ø–∞–¥–∫–æ–≤—ñ –∫–æ–ª–∏–≤–∞–Ω–Ω—è —É —Å–∫–ª–∞–¥–Ω–∏—Ö —Å–∏—Å—Ç–µ–º–∞—Ö —á–∞—Å—Ç–æ –¥–µ–º–æ–Ω—Å—Ç—Ä—É—é—Ç—å —Ä—ñ–∑–Ω–∏–π —Ä—ñ–≤–µ–Ω—å —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ —Ç–∞ —Ö–∞–æ—Ç–∏—á–Ω–æ—Å—Ç—ñ. –í —É–º–æ–≤–∞—Ö –æ–±–º–µ–∂–µ–Ω–æ—Å—Ç—ñ –¥–∞–Ω–∏—Ö —Å—Ç–∞—î –≤–∞–∂–∫–æ –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –º–µ–∂—ñ —ó—Ö –ø–µ—Ä–µ–¥–±–∞—á—É–≤–∞–Ω–æ—Å—Ç—ñ. –ê–Ω–∞–ª—ñ–∑ —Ç–∞–∫–∏—Ö —Å–∏—Å—Ç–µ–º, –ø—Ä–æ—Ü–µ—Å—ñ–≤, —â–æ –≤–∏–∑–Ω–∞—á–∞—é—Ç—å —ó—Ö –¥–∏–Ω–∞–º—ñ–∫—É, —Ç–µ–æ—Ä—ñ—è —Ö–∞–æ—Å—É —Ä–æ–∑–≥–ª—è–¥–∞–ª–∞—Å—è –≤ —Ä—ñ–∑–Ω–∏—Ö –≥–∞–ª—É–∑—è—Ö, —Ç–∞–∫–∏—Ö —è–∫ –µ–∫–æ–Ω–æ–º—ñ–∫–∞, —Ñ—ñ–Ω–∞–Ω—Å–∏, —Ñ—ñ–∑–∏–∫–∞ —Ç–∞ —ñ–Ω. –©–æ —Å—Ç–æ—Å—É—î—Ç—å—Å—è –∞–Ω–∞–ª—ñ–∑—É, –Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –¥–∏–Ω–∞–º—ñ–∫–∏ –ë—ñ—Ç–∫–æ—ó–Ω–∞, —Ç–æ –∑–Ω–∞–Ω–Ω—è –ø—Ä–æ –π–æ–≥–æ –∞–±—Å–æ–ª—é—Ç–Ω–æ –≤–∏–ø–∞–¥–∫–æ–≤—ñ —ñ, –≤–æ–¥–Ω–æ—á–∞—Å, –¥–µ—Ç–µ—Ä–º—ñ–Ω–æ–≤–∞–Ω—ñ –ø—Ä–æ—Ü–µ—Å–∏ –ø–æ—Ç–µ–Ω—Ü—ñ–π–Ω–æ –º–æ–∂—É—Ç—å –ø–æ—è—Å–Ω–∏—Ç–∏ —Ñ–ª—É–∫—Ç—É–∞—Ü—ñ—ó —á–∞—Å–æ–≤–∏—Ö —Ä—è–¥—ñ–≤ —Ä—ñ–∑–Ω–æ—ó –ø—Ä–∏—Ä–æ–¥–∏. –ó –æ–≥–ª—è–¥—É –Ω–∞ —Ñ—ñ–Ω–∞–Ω—Å–æ–≤–∏–π —Å–µ–∫—Ç–æ—Ä, –¥–æ–∫–∞–∑–∏ –¥–µ—Ç–µ—Ä–º—ñ–Ω–æ–≤–∞–Ω–æ–≥–æ —Ö–∞–æ—Å—É, –∑–Ω–∞–Ω–Ω—è —Ç–∞–∫–∏—Ö –º–æ–º–µ–Ω—Ç—ñ–≤, –∫–æ–ª–∏ –¥–≤—ñ —Å–ø–æ—á–∞—Ç–∫—É –±–ª–∏–∑—å–∫—ñ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó –ø–æ—á–∏–Ω–∞—é—Ç—å —Ä–æ–∑—Ö–æ–¥–∏—Ç–∏—Å—è, —ñ –ø–µ—Ä—ñ–æ–¥—ñ–≤, –ø—Ä–æ—Ç—è–≥–æ–º —è–∫–∏—Ö –≤–æ–Ω–∏ –∑–∞–ª–∏—à–∞—Ç–∏–º—É—Ç—å—Å—è –±–ª–∏–∑—å–∫–∏–º–∏ –æ–¥–Ω–∞ –¥–æ –æ–¥–Ω–æ—ó, –º–∞—Ç–∏–º—É—Ç—å –≤–∞–∂–ª–∏–≤–µ –∑–Ω–∞—á–µ–Ω–Ω—è –¥–ª—è —Ä–µ–≥—É–ª—è—Ç–æ—Ä—ñ–≤ —ñ —Ç—Ä–µ–π–¥–µ—Ä—ñ–≤, —è–∫—ñ —Ä–æ–∑—Ä–æ–±–ª—è—Ç–∏–º—É—Ç—å –µ—Ñ–µ–∫—Ç–∏–≤–Ω—ñ –∫–æ—Ä–æ—Ç–∫–æ—Å—Ç—Ä–æ–∫–æ–≤—ñ —Ç–æ—Ä–≥—ñ–≤–µ–ª—å–Ω—ñ —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó. –ü—Ä–æ—Ç—è–≥–æ–º –±–∞–≥–∞—Ç—å–æ—Ö —Ä–æ–∫—ñ–≤ —Ç–µ–æ—Ä—ñ—è —Ö–∞–æ—Å—É –Ω–∞–¥–∞–≤–∞–ª–∞ –ø—ñ–¥—Ö–æ–¥–∏ –¥–æ –≤–∏–≤—á–µ–Ω–Ω—è –¥–µ—è–∫–∏—Ö —Ü—ñ–∫–∞–≤–∏—Ö –≤–ª–∞—Å—Ç–∏–≤–æ—Å—Ç–µ–π —á–∞—Å–æ–≤–∏—Ö —Ä—è–¥—ñ–≤. –ù–∞–π–±—ñ–ª—å—à –ø–æ—à–∏—Ä–µ–Ω–∏–º–∏ —î: –∫–æ—Ä–µ–ª—è—Ü—ñ–π–Ω–∞ —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å, BDS —Ç–µ—Å—Ç, –µ–Ω—Ç—Ä–æ–ø—ñ—è –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞, –ø–æ–∫–∞–∑–Ω–∏–∫–∏ –õ—è–ø—É–Ω–æ–≤–∞ —Ç–æ—â–æ.

–ó–∞—Å—Ç–æ—Å—É–≤–∞–≤—à–∏ –¥–æ —á–∞—Å–æ–≤–æ–≥–æ —Ä—è–¥—É –¥–æ—Å–ª—ñ–¥–∂—É–≤–∞–Ω–æ—ó —Å–∏—Å—Ç–µ–º–∏ –ø—ñ–¥—Ö—ñ–¥ –∫–æ–≤–∑–Ω–æ–≥–æ –≤—ñ–∫–Ω–∞ —Ç–∞ –ø–æ–∫–∞–∑–Ω–∏–∫—ñ–≤ –õ—è–ø—É–Ω–æ–≤–∞, –º–∏ –≤—ñ–¥–∑–µ—Ä–∫–∞–ª—é–≤–∞—Ç–∏–º–µ–º–æ –π–æ–≥–æ –ø–µ—Ä–µ—Ö—ñ–¥ –º—ñ–∂ —Ö–∞–æ—Ç–∏—á–Ω–æ—é —Ç–∞ –Ω–µ—Ö–∞–æ—Ç–∏—á–Ω–æ—é –ø–æ–≤–µ–¥—ñ–Ω–∫–æ—é.

### –ü–æ–∫–∞–∑–Ω–∏–∫–∏ –õ—è–ø—É–Ω–æ–≤–∞

–ï–≤–æ–ª—é—Ü—ñ—è —Å–∏—Å—Ç–µ–º–∏ –¥–µ–º–æ–Ω—Å—Ç—Ä—É—î *—á—É—Ç–ª–∏–≤—É –∑–∞–ª–µ–∂–Ω—ñ—Å—Ç—å* –≤—ñ–¥ *–ø–æ—á–∞—Ç–∫–æ–≤–∏—Ö —É–º–æ–≤*. –¶–µ –æ–∑–Ω–∞—á–∞—î, —â–æ —Å–ø–æ—á–∞—Ç–∫—É –±–ª–∏–∑—å–∫—ñ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó, —è–∫—ñ —Ä–æ–∑–≤–∏–≤–∞—é—Ç—å—Å—è, –º–æ–∂—É—Ç—å —à–≤–∏–¥–∫–æ –≤—ñ–¥—Ö–∏–ª—è—Ç–∏—Å—è –æ–¥–Ω–∞ –≤—ñ–¥ –æ–¥–Ω–æ—ó —ñ –º–∞—Ç–∏ –∞–±—Å–æ–ª—é—Ç–Ω–æ —Ä—ñ–∑–Ω—ñ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏. –í—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ, –ø—Ä–∏ –º–∞–ª–∏—Ö –Ω–µ–≤–∏–∑–Ω–∞—á–µ–Ω–æ—Å—Ç—è—Ö, —è–∫—ñ –Ω–∞–¥–∑–≤–∏—á–∞–π–Ω–æ —à–≤–∏–¥–∫–æ –ø–æ—Å–∏–ª—é—é—Ç—å—Å—è, –¥–æ–≤–≥–æ—Å—Ç—Ä–æ–∫–æ–≤—ñ –ø—Ä–æ–≥–Ω–æ–∑–∏ –≤–∏—è–≤–ª—è—é—Ç—å—Å—è –Ω–µ–º–æ–∂–ª–∏–≤–∏–º–∏. –ó —ñ–Ω—à–æ–≥–æ –±–æ–∫—É, –≤ —Å–∏—Å—Ç–µ–º—ñ –∑ —Ç–æ—á–∫–∞–º–∏ —Ç—è–∂—ñ–Ω–Ω—è –∞–±–æ —Å—Ç–∞–±—ñ–ª—å–Ω–∏–º–∏ —Ç–æ—á–∫–∞–º–∏ –≤—ñ–¥—Å—Ç–∞–Ω—å –º—ñ–∂ –Ω–∏–º–∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–æ –∑–º–µ–Ω—à—É—î—Ç—å—Å—è –∑ —á–∞—Å–æ–º –∞–±–æ –∑ –∫—ñ–ª—å–∫—ñ—Å—Ç—é —Ç–æ—á–æ–∫, —è–∫—ñ –º–∞—é—Ç—å —Ç–µ–Ω–¥–µ–Ω—Ü—ñ—é –¥–æ –∑–±–ª–∏–∂–µ–Ω–Ω—è.

–©–æ–± –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç–∏ —ñ–¥–µ—é –±—ñ–ª—å—à —Ç–æ—á–Ω–æ, —Ä–æ–∑–≥–ª—è–Ω–µ–º–æ –¥–≤—ñ –ø–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó --- $x(t)$ —Ç–∞ –Ω–∞–±–ª–∏–∂—á–æ–≥–æ —Å—É—Å—ñ–¥–∞ —Ü—ñ—î—ó —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó –∑ –Ω–µ–≤–µ–ª–∏–∫–∏–º –∑–º—ñ—â–µ–Ω–Ω—è–º, $x(t) + \delta(t)$, –¥–µ $\delta(t)$ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—î —Å–æ–±–æ—é –∫—Ä–∏—Ö—ñ—Ç–Ω–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è –≤ —á–∞—Å—ñ $t$, —è–∫ –ø–æ–∫–∞–∑–∞–Ω–æ –Ω–∞ [@fig-sensitivity].

–†–æ–∑–±—ñ–∂–Ω—ñ—Å—Ç—å –¥–≤–æ—Ö –ø–æ—á–∞—Ç–∫–æ–≤–æ –±–ª–∏–∑—å–∫–∏—Ö —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ–π —É –¥–∏–Ω–∞–º—ñ—á–Ω—ñ–π —Å–∏—Å—Ç–µ–º—ñ

–ö–æ–ª–∏ –¥–∏–Ω–∞–º—ñ–∫–∞ –¥–≤–æ—Ö –ø–æ—á–∞—Ç–∫–æ–≤–æ –±–ª–∏–∑—å–∫–∏—Ö —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ–π –ø–æ—Ä—É—à—É—é—Ç—å—Å—è –ø–µ–≤–Ω–æ—é –ø–æ–¥—ñ—î—é, –≤–æ–Ω–∏ –ø–æ—á–∏–Ω–∞—é—Ç—å —Ä–æ–∑—Ö–æ–¥–∏—Ç–∏—Å—è, —ñ –≤—ñ–¥—Å—Ç–∞–Ω—å –º—ñ–∂ –Ω–∏–º–∏ –∑—Ä–æ—Å—Ç–∞—î –∑–∞ –µ–∫—Å–ø–æ–Ω–µ–Ω—Ü—ñ–∞–ª—å–Ω–∏–º –∑–∞–∫–æ–Ω–æ–º:

$$
\| \delta(t) \| \approx \| \delta(0) \| \exp(\lambda t), 
$$ {#eq-9-1}

–¥–µ $\lambda$ –ø–æ–∑–Ω–∞—á–∞—î *–ø–æ–∫–∞–∑–Ω–∏–∫ –õ—è–ø—É–Ω–æ–≤–∞* (–ü–õ); $\delta(t)$ --- –≤—ñ–¥—Å—Ç–∞–Ω—å –º—ñ–∂ —Ç–æ—á–∫–æ—é —â–æ —Ä–æ–∑–≥–ª—è–¥–∞—î—Ç—å—Å—è —Ç–∞ —ó—ó –Ω–∞–±–ª–∏–∂—á–∏–º —Å—É—Å—ñ–¥–æ–º –ø—ñ—Å–ª—è —á–∞—Å—É $t$ (–∞–±–æ –ø—ñ—Å–ª—è $t$ —ñ—Ç–µ—Ä–∞—Ü—ñ–π); $\delta(0)$ --- —Ü–µ –ø–æ—á–∞—Ç–∫–æ–≤–∞ –≤—ñ–¥—Å—Ç–∞–Ω—å –º—ñ–∂ —Ç–æ—á–∫–æ—é —â–æ —Ä–æ–∑–≥–ª—è–¥–∞—î—Ç—å—Å—è —Ç–∞ —ó—ó –Ω–∞–π–±–ª–∏–∂—á–∏–º —Å—É—Å—ñ–¥–æ–º —É –ø–æ—á–∞—Ç–∫–æ–≤–∏–π –º–æ–º–µ–Ω—Ç —á–∞—Å—É ($t=0$). 

–ü–õ —î –º—ñ—Ä–æ—é —à–≤–∏–¥–∫–æ—Å—Ç—ñ –µ–∫—Å–ø–æ–Ω–µ–Ω—Ü—ñ–∞–ª—å–Ω–æ—ó —Ä–æ–∑–±—ñ–∂–Ω–æ—Å—Ç—ñ –±–ª–∏–∑—å–∫–∏—Ö –æ–¥–∏–Ω –¥–æ –æ–¥–Ω–æ–≥–æ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ–π —É —Ñ–∞–∑–æ–≤–æ–º—É –ø—Ä–æ—Å—Ç–æ—Ä—ñ –¥–∏–Ω–∞–º—ñ—á–Ω–æ—ó —Å–∏—Å—Ç–µ–º–∏. –Ü–Ω—à–∏–º–∏ —Å–ª–æ–≤–∞–º–∏, –ü–õ –ø–æ–∫–∞–∑—É—î, –Ω–∞—Å–∫—ñ–ª—å–∫–∏ —à–≤–∏–¥–∫–æ –∑–±–ª–∏–∂—É—é—Ç—å—Å—è –∞–±–æ —Ä–æ–∑—Ö–æ–¥—è—Ç—å—Å—è —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó, —è–∫—ñ –ø–æ—á–∏–Ω–∞—é—Ç—å—Å—è –±–ª–∏–∑—å–∫–æ –æ–¥–Ω–∞ –≤—ñ–¥ –æ–¥–Ω–æ—ó, –≤–∏–º—ñ—Ä—é—é—á–∏ —Å—Ç—É–ø—ñ–Ω—å —Ö–∞–æ—Å—É –≤ —Å–∏—Å—Ç–µ–º—ñ.

–£ —Ç–∏—Ö –≤–∏–ø–∞–¥–∫–∞—Ö, –∫–æ–ª–∏ –Ω–∞—à–∞ —Å–∏—Å—Ç–µ–º–∞ $n$-–≤–∏–º—ñ—Ä–Ω–∞, –º–∏ –º–∞—î–º–æ —Å—Ç—ñ–ª—å–∫–∏ –ü–õ, —Å–∫—ñ–ª—å–∫–∏ –≤–∏–º—ñ—Ä—ñ–≤ —É –Ω—ñ–π. –î–ª—è —ó—Ö –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è —Ä–æ–∑–≥–ª—è–Ω–µ–º–æ –µ–≤–æ–ª—é—Ü—ñ—é –Ω–µ—Å–∫—ñ–Ω—á–µ–Ω–Ω–æ –º–∞–ª–æ—ó —Å—Ñ–µ—Ä–∏, —â–æ –∑–∞–∑–Ω–∞–ª–∞ –∑–±—É—Ä–µ–Ω—å –∑–∞ —Ä—ñ–∑–Ω–∏–º–∏ –æ—Å—è–º–∏ (–∑–∞–∑–Ω–∞—á–∞–ª–∞ –ø–æ—á–∞—Ç–∫–æ–≤–∏—Ö —É–º–æ–≤). –í–∏–∑–Ω–∞—á–∏–≤—à–∏  –≤–µ–ª–∏—á–∏–Ω—É –∑–±—É—Ä–µ–Ω–Ω—è –ø–æ –≤—ñ—Å—ñ $i$ —è–∫ $\delta_i(t)$, –æ—Ç—Ä–∏–º–∞—î–º–æ $n$ –ø–æ–∫–∞–∑–Ω–∏–∫—ñ–≤ –õ—è–ø—É–Ω–æ–≤–∞, —â–æ –º–∞—é—Ç—å –≤–∏–¥

$$
\| \delta_i(t) \| \approx \| \delta_i(0) \| \exp(\lambda_i t), \; \text{–¥–ª—è} \; i=1,...,n.
$$ {#eq-9-2}

–î–ª—è –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è —Ç–æ–≥–æ, —á–∏ —î —Ä—É—Ö –ø–µ—Ä—ñ–æ–¥–∏—á–Ω–∏–º –∞–±–æ —Ö–∞–æ—Ç–∏—á–Ω–∏–º, –æ—Å–æ–±–ª–∏–≤–æ –¥–ª—è –≤–µ–ª–∏–∫–∏—Ö $t$, —Ä–µ–∫–æ–º–µ–Ω–¥—É—î—Ç—å—Å—è —Ä–æ–∑–≥–ª—è–¥–∞—Ç–∏ –≤–Ω–µ—Å–æ–∫ —Å–∏—Å—Ç–µ–º–∏ –≤ –Ω–∞–π–±—ñ–ª—å—à–∏–π –ø–æ–∫–∞–∑–Ω–∏–∫ –õ—è–ø—É–Ω–æ–≤–∞ (–ù–ü–õ), –æ—Å–∫—ñ–ª—å–∫–∏ –¥—ñ–∞–º–µ—Ç—Ä $n$-—Ä–æ–∑–º—ñ—Ä–Ω–æ–≥–æ –µ–ª—ñ–ø—Å–æ—ó–¥–∞ –ø–æ—á–∏–Ω–∞—î –∑–∞–ª–µ–∂–∞—Ç–∏ –≤—ñ–¥ –Ω–µ—ó. –°–∞–º–µ –ù–ü–õ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –∫—ñ–ª—å–∫—ñ—Å–Ω–æ—ó –æ—Ü—ñ–Ω–∫–∏ –ø–µ—Ä–µ–¥–±–∞—á—É–≤–∞–Ω–æ—Å—Ç—ñ —Å–∏—Å—Ç–µ–º, –æ—Å–∫—ñ–ª—å–∫–∏ –µ–∫—Å–ø–æ–Ω–µ–Ω—Ü—ñ–∞–ª—å–Ω–∞ —Ä–æ–∑–±—ñ–∂–Ω—ñ—Å—Ç—å –æ–∑–Ω–∞—á–∞—î, —â–æ –≤ —Å–∏—Å—Ç–µ–º—ñ, –¥–µ –ø–æ—á–∞—Ç–∫–æ–≤–µ –∑–±—É—Ä–µ–Ω–Ω—è –±—É–ª–æ –Ω–µ—Å–∫—ñ–Ω—á–µ–Ω–Ω–æ –º–∞–ª–∏–º, –ø–æ—á–∏–Ω–∞—é—Ç—å—Å—è –≤—Ç—Ä–∞—Ç–∏ –ø–µ—Ä–µ–¥–±–∞—á—É–≤–∞–Ω–Ω–æ—Å—Ç—ñ. –û–¥–Ω–∞–∫ —Å–ª—ñ–¥ –∑–∞–∑–Ω–∞—á–∏—Ç–∏, —â–æ —ñ–Ω—à—ñ –µ–∫—Å–ø–æ–Ω–µ–Ω—Ç–∏ —Ç–∞–∫–æ–∂ –º—ñ—Å—Ç—è—Ç—å –≤–∞–∂–ª–∏–≤—É —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—é –ø—Ä–æ —Å—Ç—ñ–π–∫—ñ—Å—Ç—å —Å–∏—Å—Ç–µ–º–∏, –≤ —Ç–æ–º—É —á–∏—Å–ª—ñ –ø—Ä–æ –Ω–∞–ø—Ä—è–º–∫–∏ –∑–±—ñ–∂–Ω–æ—Å—Ç—ñ —Ç–∞ —Ä–æ–∑–±—ñ–∂–Ω–æ—Å—Ç—ñ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ–π.

–Ü—Å–Ω—É–≤–∞–Ω–Ω—è –ø—Ä–∏–Ω–∞–π–º–Ω—ñ –æ–¥–Ω–æ–≥–æ –ø–æ–∑–∏—Ç–∏–≤–Ω–æ–≥–æ –ü–õ –∑–∞–∑–≤–∏—á–∞–π —Ä–æ–∑–≥–ª—è–¥–∞—î—Ç—å—Å—è —è–∫ —Å–∏–ª—å–Ω–∏–π —ñ–Ω–¥–∏–∫–∞—Ç–æ—Ä —Ö–∞–æ—Å—É. –ü–æ–∑–∏—Ç–∏–≤–Ω–∏–π –ü–õ –æ–∑–Ω–∞—á–∞—î, —â–æ –ø–æ—á–∞—Ç–∫–æ–≤–æ –±–ª–∏–∑—å–∫—ñ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó —É —Ñ–∞–∑–æ–≤–æ–º—É –ø—Ä–æ—Å—Ç–æ—Ä—ñ, —á—É—Ç–ª–∏–≤—ñ –¥–æ –ø–æ—á–∞—Ç–∫–æ–≤–∏—Ö —É–º–æ–≤ —ñ —Ä–æ–∑—Ö–æ–¥—è—Ç—å—Å—è –µ–∫—Å–ø–æ–Ω–µ–Ω—Ü—ñ–∞–ª—å–Ω–æ —à–≤–∏–¥–∫–æ, —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É—é—Ç—å —Ö–∞–æ—Ç–∏—á–Ω—É –ø–æ–≤–µ–¥—ñ–Ω–∫—É —Å–∏—Å—Ç–µ–º–∏. –ù–µ–≥–∞—Ç–∏–≤–Ω–∏–π –ü–õ –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—î –≤–∏–ø–∞–¥–∫–∞–º, –∫–æ–ª–∏ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó –∑–∞–ª–∏—à–∞—é—Ç—å—Å—è –±–ª–∏–∑—å–∫–∏–º–∏ –æ–¥–Ω–∞ –¥–æ –æ–¥–Ω–æ—ó, –∞–ª–µ —Ü–µ –Ω–µ –æ–±–æ–≤'—è–∑–∫–æ–≤–æ –æ–∑–Ω–∞—á–∞—î —Å—Ç–∞–±—ñ–ª—å–Ω—ñ—Å—Ç—å, —ñ –º–∏ –ø–æ–≤–∏–Ω–Ω—ñ –¥–æ—Å–ª—ñ–¥–∏—Ç–∏ –Ω–∞—à—É —Å–∏—Å—Ç–µ–º—É –±—ñ–ª—å—à –¥–µ—Ç–∞–ª—å–Ω–æ. –ù—É–ª—å–æ–≤—ñ –∞–±–æ –¥—É–∂–µ –±–ª–∏–∑—å–∫—ñ –¥–æ –Ω—É–ª—è –µ–∫—Å–ø–æ–Ω–µ–Ω—Ç–∏ –≤–∫–∞–∑—É—é—Ç—å –Ω–∞ —Ç–µ, —â–æ –∑–±—É—Ä–µ–Ω–Ω—è, –≤–Ω–µ—Å–µ–Ω—ñ –≤–∑–¥–æ–≤–∂ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó, –Ω–µ —Ä–æ–∑—Ö–æ–¥—è—Ç—å—Å—è —ñ –Ω–µ –∑–±–ª–∏–∂—É—é—Ç—å—Å—è.

–£ –∑–≤'—è–∑–∫—É –∑ –≤–µ–ª–∏–∫–æ—é –∑–∞—Ü—ñ–∫–∞–≤–ª–µ–Ω—ñ—Å—Ç—é –≤ –ü–õ, –∑'—è–≤–ª—è—î—Ç—å—Å—è –≤—Å–µ –±—ñ–ª—å—à–µ –ø—Ä–æ–ø–æ–∑–∏—Ü—ñ–π —â–æ–¥–æ —ó—Ö —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É. –ù–∞ –∂–∞–ª—å, –¥–æ—Å—ñ –Ω–µ –æ—Ç—Ä–∏–º–∞–Ω–æ –∑–∞–≥–∞–ª—å–Ω–æ–ø—Ä–∏–π–Ω—è—Ç–æ–≥–æ —Ç–∞ —É–Ω—ñ–≤–µ—Ä—Å–∞–ª—å–Ω–æ–≥–æ –º–µ—Ç–æ–¥—É –æ—Ü—ñ–Ω–∫–∏ –≤—Å—å–æ–≥–æ —Å–ø–µ–∫—Ç—Ä—É –ø–æ–∫–∞–∑–Ω–∏–∫—ñ–≤ –õ—è–ø—É–Ω–æ–≤–∞ –∑–∞ –∑–Ω–∞—á–µ–Ω–Ω—è–º–∏ —á–∞—Å–æ–≤–æ–≥–æ —Ä—è–¥—É. –û–¥–Ω—ñ –∑ –Ω–∞–π–±—ñ–ª—å—à –ø–æ—à–∏—Ä–µ–Ω–∏—Ö —ñ –ø–æ–ø—É–ª—è—Ä–Ω–∏—Ö –∞–ª–≥–æ—Ä–∏—Ç–º—ñ–≤ –±—É–ª–∏ –∑–∞—Å—Ç–æ—Å–æ–≤–∞–Ω—ñ –í–æ–ª—å—Ñ–æ–º —Ç–∞ —ñ–Ω., –°–∞–Ω–æ —ñ –°–∞–≤–∞–¥–æ—é, –∞ –ø—ñ–∑–Ω—ñ—à–µ –≤–¥–æ—Å–∫–æ–Ω–∞–ª–µ–Ω—ñ –ï–∫–º–∞–Ω–æ–º, –†–æ–∑–µ–Ω—à—Ç–µ–π–Ω–æ–º, –ü–∞—Ä–ª—ñ—Ü–æ–º, –ë–∞–ª—å—Ü–µ—Ä–∂–∞–∫–æ–º —Ç–æ—â–æ.

–¢–µ–æ—Ä—ñ—è —Ö–∞–æ—Å—É —Ç–∞ —ó—ó —ñ–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–∞—Ä—ñ–π –∑–∞–ª–∏—à–∞—é—Ç—å—Å—è –≤–µ–ª–∏—á–µ–∑–Ω–∏–º –≤–∏–∫–ª–∏–∫–æ–º –¥–ª—è –¥–æ—Å–ª—ñ–¥–Ω–∏–∫—ñ–≤ —Ä—ñ–∑–Ω–∏—Ö –≥–∞–ª—É–∑–µ–π –Ω–∞—É–∫–∏. –£ —Å–≤—ñ—Ç—ñ –ø–æ–∫–∞–∑–Ω–∏–∫—ñ–≤ –õ—è–ø—É–Ω–æ–≤–∞ –∑–±–µ—Ä—ñ–≥–∞—î—Ç—å—Å—è –∑—Ä–æ—Å—Ç–∞—é—á–∏–π —ñ–Ω—Ç–µ—Ä–µ—Å –¥–æ —ó—Ö –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è, —á–∏—Å–µ–ª—å–Ω–∏—Ö –º–µ—Ç–æ–¥—ñ–≤ —Ç–∞ –∑–∞—Å—Ç–æ—Å—É–≤–∞–Ω–Ω—è –¥–æ —Ä—ñ–∑–Ω–∏—Ö —Å–∫–ª–∞–¥–Ω–∏—Ö —Å–∏—Å—Ç–µ–º. –ü—ñ–¥—Å—É–º–æ–≤—É—é—á–∏, –°–ü–õ –¥–æ–∑–≤–æ–ª—è—î –≤—Å—Ç–∞–Ω–æ–≤–∏—Ç–∏:
- –æ–±–ª–∞—Å—Ç—å —á—É—Ç–ª–∏–≤–æ—Å—Ç—ñ –¥–æ –ø–æ—á–∞—Ç–∫–æ–≤–∏—Ö —É–º–æ–≤;
- –æ–±–ª–∞—Å—Ç—å —Ö–∞–æ—Å—É;
- –æ–±–ª–∞—Å—Ç—å —Å—Ç–∞–±—ñ–ª—å–Ω–æ—Å—Ç—ñ.

#### –ú–µ—Ç–æ–¥ –ï–∫–º–∞–Ω–∞

–ü–æ-–ø–µ—Ä—à–µ, –∑–≥—ñ–¥–Ω–æ –∑ –ø—ñ–¥—Ö–æ–¥–æ–º –ï–∫–º–∞–Ω–∞ —Ç–∞ —ñ–Ω., –º–∏ –ø–æ–≤–∏–Ω–Ω—ñ —Ä–µ–∫–æ–Ω—Å—Ç—Ä—É—é–≤–∞—Ç–∏ –¥–∏–Ω–∞–º—ñ–∫—É –∞—Ç—Ä–∞–∫—Ç–æ—Ä–∞ –∑ –æ–¥–Ω–æ–≥–æ —á–∞—Å–æ–≤–æ–≥–æ —Ä—è–¥—É $\{x(i) \,|\, i=1,...N \}$ –∑ —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—é –≤–∫–ª–∞–¥–µ–Ω—å $d_E$, —ñ –ø—ñ—Å–ª—è —Ü—å–æ–≥–æ –ø–æ–±—É–¥—É–≤–∞—Ç–∏ $d_E$-–≤–∏–º—ñ—Ä–Ω—É –æ—Ä–±—ñ—Ç—É, —â–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—î —á–∞—Å–æ–≤—É –µ–≤–æ–ª—é—Ü—ñ—é

$$
\vec{X}(i) = \left[ x(i), x(i+1), ..., x(i+(d_E - 1)) \right], \; \text{–¥–ª—è} \; i=1,...,N-d_E+1. 
$$

–î–∞–ª—ñ, –º–∏ –º–∞—î–º–æ –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –Ω–∞–π–±–ª–∏–∂—á–∏ –¥–æ $\vec{X}(i)$ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó:

$$
\| \vec{X}(i) - \vec{X}(j) \| = \max_{0\leq \alpha \leq d_E-1} \left| x(i+\alpha) - x(j+\alpha) \right|.   
$$ {#eq-9-3}

–ú–∏ —Å–æ—Ä—Ç—É—î–º–æ $x(i)$ —Ç–∞–∫, —â–æ–± $x(\Pi(1)) \leq x(\Pi(2)) \leq ... \leq x(\Pi(N))$ —ñ –∑–±–µ—Ä—ñ–≥–∞—î–º–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫—É $\Pi$ —Ç–∞ —ó—ó –∑–≤–æ—Ä–æ—Ç–Ω—é –≤–µ—Ä—Å—ñ—é $\Pi^-1$. –î–∞–ª—ñ –º–∏ –Ω–∞–º–∞–≥–∞—î–º–æ—Å—å –∑–Ω–∞–π—Ç–∏ —Å—É—Å—ñ–¥—ñ–≤ $x(i)$, –ø–µ—Ä–µ–≥–ª—è–¥–∞—é—á–∏ $k=\Pi^{-1}(i)$ —ñ —Å–∫–∞–Ω—É—î–º–æ $x(\Pi(s))$ –ø—Ä–∏ $s=k+1, k+2, ... \text{—ñ} k-1, k-2, ... $ –¥–æ —Ç–∏—Ö –ø—ñ—Ä, –¥–æ –ø–æ–∫–∏ –Ω–µ –≤–∏–∫–æ–Ω–∞—î—Ç—å—Å—è —É–º–æ–≤–∞ $x(\Pi(ùë†)) - x(i) > r$. –î–ª—è –≤–∏–±—Ä–∞–Ω–æ—ó —Ä–æ–∑–º—ñ—Ä–Ω–æ—Å—Ç—ñ –≤–∫–ª–∞–¥–µ–Ω–Ω—è $d_E > 1$ –≤–∏–±–∏—Ä–∞—î–º–æ –∑–Ω–∞—á–µ–Ω–Ω—è $s$, –¥–ª—è —è–∫–æ–≥–æ –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –Ω–∞—Å—Ç—É–ø–Ω–∞ —É–º–æ–≤–∞

$$
\left| x(\Pi(s) + \alpha) - x(j + \alpha) \right| \leq r, \; \text{–¥–ª—è} \; \alpha=0,1,...,d_E-1. 
$$

–ü—ñ—Å–ª—è —Ç–æ–≥–æ, —è–∫ –º–∏ —Ä–µ–∫–æ–Ω—Å—Ç—Ä—É—é–≤–∞–ª–∏ –Ω–∞—à—É —Å–∏—Å—Ç–µ–º—É —É —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å $d_E$, –Ω–∞–º –ø–æ—Ç—Ä—ñ–±–Ω–æ –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –º–∞—Ç—Ä–∏—Ü—é $M_i$ —Ä–æ–∑–º—ñ—Ä–Ω–æ—Å—Ç—ñ $d_E \times d_E$, —è–∫–∞ –æ–ø–∏—Å—É–≤–∞—Ç–∏–º–µ —á–∞—Å–æ–≤—É –µ–≤–æ–ª—é—Ü—ñ—é –≤–µ–∫—Ç–æ—Ä—ñ–≤, —â–æ –æ—Ç–æ—á—É—é—Ç—å —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—é $\vec{X}(i)$, —ñ —Ç–µ, —è–∫ –≤–æ–Ω–∏ –≤—ñ–¥–æ–±—Ä–∞–∂–∞—é—Ç—å—Å—è –Ω–∞ —Å—Ç–∞–Ω $\vec{X}(i+1)$. –ú–∞—Ç—Ä–∏—Ü—è $M_i$ –æ—Ç—Ä–∏–º—É—î—Ç—å—Å—è —à–ª—è—Ö–æ–º –ø–æ—à—É–∫—É —Å—É—Å—ñ–¥—ñ–≤

$$
M_i(\vec{X}(i) - \vec{X}(j)) \approx \vec{X}(i+1) - \vec{X}(j+1). 
$$ {#eq-9-4}

–ü—Ä–æ—Ç–µ –≤–µ–∫—Ç–æ—Ä–∏ $\vec{X}(i)-\vec{X}(j)$ –º–æ–∂—É—Ç—å —ñ –Ω–µ –ø–æ–∫—Ä–∏–≤–∞—Ç–∏ $\mathbb{R}^{d_E}$. –£ —Ü—å–æ–º—É –≤–∏–ø–∞–¥–∫—É —Ç–∞–∫–∞ –Ω–µ–≤–∏–∑–Ω–∞—á–µ–Ω—ñ—Å—Ç—å –º–æ–∂–µ –ø—Ä–∏–∑–≤–µ—Å—Ç–∏ –¥–æ —Ö–∏–±–Ω–∏—Ö –µ–∫—Å–ø–æ–Ω–µ–Ω—Ç, —è–∫—ñ –º–æ–∂—É—Ç—å –∑—ñ–ø—Å—É–≤–∞—Ç–∏ –∞–Ω–∞–ª—ñ–∑. –î–ª—è –ø–æ–¥–æ–ª–∞–Ω–Ω—è —Ç–∞–∫–∏—Ö –ø–µ—Ä–µ—à–∫–æ–¥ –ø—Ä–æ–µ–∫—Ü—ñ—è —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ–π –≤–∏–∑–Ω–∞—á–∞—î—Ç—å—Å—è –Ω–∞ –ø—ñ–¥–ø—Ä–æ—Å—Ç—ñ—Ä —Ä–æ–∑–º—ñ—Ä–Ω–æ—Å—Ç—ñ $d_M \leq d_E$. –¢–∞–∫–∏–º —á–∏–Ω–æ–º, –ø—Ä–æ—Å—Ç—ñ—Ä, –Ω–∞ —è–∫–æ–º—É –≤—ñ–¥–±—É–≤–∞—î—Ç—å—Å—è –¥–∏–Ω–∞–º—ñ–∫–∞, –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—î –ª–æ–∫–∞–ª—å–Ω—ñ–π —Ä–æ–∑–º—ñ—Ä–Ω–æ—Å—Ç—ñ $d_M$, –¥–µ $d_E$ –º–∞—î –±—É—Ç–∏ –¥–µ—â–æ –±—ñ–ª—å—à–∏–º –∑–∞ $d_M$, —â–æ–± —É–Ω–∏–∫–Ω—É—Ç–∏ –Ω–∞—è–≤–Ω–æ—Å—Ç—ñ —Ö–∏–±–Ω–∏—Ö —Å—É—Å—ñ–¥—ñ–≤. –ó–≤—ñ–¥—Å–∏ –≤–∏–ø–ª–∏–≤–∞—î, —â–æ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—è $\vec{X}(i)$ –∞—Å–æ—Ü—ñ—é—î—Ç—å—Å—è –∑ $d_M$-–≤–∏–º—ñ—Ä–Ω–∏–º –≤–µ–∫—Ç–æ—Ä–æ–º

$$
\vec{X}(i) = \left[ x(i), x(i+\tau), ..., x(i+(d_M - 1)\tau) \right] = \left[ x(i), x(i+\tau), ..., x(i+d_E - 1) \right], 
$$ {#eq-9-5}

–¥–µ $\tau=(d_E-1)/(d_M-1)$. –ö–æ–ª–∏ $\tau>1$, —É–º–æ–≤–∞ ([-@eq-9-4]) –∑–∞–º—ñ–Ω—é—î—Ç—å—Å—è –Ω–∞—Å—Ç—É–ø–Ω–∏–º –≤–∏—Ä–∞–∑–æ–º:

$$
M_i(\vec{X}(i) - \vec{X}(j)) \approx \vec{X}(i+\tau) - \vec{X}(j+\tau). 
$$ {#eq-9-6}

–ú–∞—Ç—Ä–∏—Ü—è $M_i$ –ø–æ—Ç—ñ–º –≤–∏–∑–Ω–∞—á–∞—î—Ç—å—Å—è –º–µ—Ç–æ–¥–æ–º –Ω–∞–π–º–µ–Ω—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç—ñ–≤. –û—Å—Ç–∞–Ω–Ω—ñ–º –∫—Ä–æ–∫–æ–º —î QR –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü—ñ—è, –¥–ª—è –∑–Ω–∞—Ö–æ–¥–∂–µ–Ω–Ω—è –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω–∏—Ö –º–∞—Ç—Ä–∏—Ü—å $Q_i$ —ñ –≤–µ—Ä—Ö–Ω—ñ—Ö —Ç—Ä–∏–∫—É—Ç–Ω–∏—Ö –º–∞—Ç—Ä–∏—Ü—å $R_i$ –ø—Ä–∏ —è–∫–∏—Ö 

$$
M_{1+i\tau}Q_i = Q_{i+1}R_{i+1}, \; \text{–¥–ª—è} \; i=0,1,2,... .
$$

–Ø–∫ –±—É–ª–æ –∑–∞–ø—Ä–æ–ø–æ–Ω–æ–≤–∞–Ω–æ –ï–∫–º–∞–Ω–æ–º, –¥–ª—è –∑–Ω–∞—Ö–æ–¥–∂–µ–Ω–Ω—è $d_M$ –ø–æ–∫–∞–∑–Ω–∏–∫—ñ–≤ –õ—è–ø—É–Ω–æ–≤–∞, –∑–Ω–∞—é—á–∏ $K$ –∫—ñ–ª—å–∫—ñ—Å—Ç—å —Ç–æ—á–æ–∫ –Ω–∞ –∞—Ç—Ä–∞–∫—Ç–æ—Ä—ñ, –¥—ñ–∞–≥–æ–Ω–∞–ª—å–Ω—ñ –≤–ª–∞—Å–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—ñ $R_i$ —Ç–∞ –∫—Ä–æ–∫ –¥–∏—Å–∫—Ä–µ—Ç–∏–∑–∞—Ü—ñ—ó $\Delta t$, –º–æ–∂–Ω–∞ –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –Ω–∞—Å—Ç—É–ø–Ω–µ —Ä—ñ–≤–Ω—è–Ω–Ω—è –¥–ª—è –∑–Ω–∞—Ö–æ–¥–∂–µ–Ω–Ω—è $k$-–≥–æ –ü–õ:

$$
\lambda_k = \frac{1}{\Delta t}\frac{1}{\tau}\frac{1}{K}\sum_{i=0}^{K-1}\ln{(R_i)_{kk}}. 
$$

–¢–∞–∫–∏–º —á–∏–Ω–æ–º, –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –ª—ñ–Ω–µ–∞—Ä–∏–∑–∞—Ü—ñ—ó —ñ–∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º –¥—ñ–∞–≥–æ–Ω–∞–ª—å–Ω–∏—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –∑ QR-—Ä–æ–∑–∫–ª–∞–¥—É –º–∏ –º–æ–∂–µ–º–æ
–æ–±—á–∏—Å–ª–∏—Ç–∏ –ø–æ–∫–∞–∑–Ω–∏–∫–∏ –õ—è–ø—É–Ω–æ–≤–∞.

#### –ú–µ—Ç–æ–¥ –†–æ–∑–µ–Ω—à—Ç–µ–π–Ω–∞

–ê–ª–≥–æ—Ä–∏—Ç–º –†–æ–∑–µ–Ω—à—Ç–µ–π–Ω–∞ [@ROSENSTEIN1993117] –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î –º–µ—Ç–æ–¥ —Ä–µ–∫–æ–Ω—Å—Ç—Ä—É–∫—Ü—ñ—ó –≤–∫–ª–∞–¥–µ–Ω—å —ñ–∑ —á–∞—Å–æ–≤–æ—é –∑–∞—Ç—Ä–∏–º–∫–æ—é, —è–∫–∏–π —Ä–µ–∫–æ–Ω—Å—Ç—Ä—É—é—î –Ω–∞–π–≤–∞–∂–ª–∏–≤—ñ—à—ñ –æ—Å–æ–±–ª–∏–≤–æ—Å—Ç—ñ –±–∞–≥–∞—Ç–æ–≤–∏–º—ñ—Ä–Ω–æ–≥–æ –∞—Ç—Ä–∞–∫—Ç–æ—Ä–∞ –≤ –æ–¥–∏–Ω –æ–¥–Ω–æ–≤–∏–º—ñ—Ä–Ω–∏–π —á–∞—Å–æ–≤–∏–π —Ä—è–¥ –¥–µ—è–∫–æ–≥–æ —Å–∫—ñ–Ω—á–µ–Ω–Ω–æ–≥–æ —Ä–æ–∑–º—ñ—Ä—É $N$. –î–ª—è —á–∞—Å–æ–≤–æ–≥–æ —Ä—è–¥—É –∫–æ–∂–µ–Ω –≤–µ–∫—Ç–æ—Ä $\vec{X}(i)$ –±—É–¥–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–π –ø–æ–¥—ñ–±–Ω–æ –¥–æ –≤–µ–∫—Ç–æ—Ä–∞ ([-@eq-9-5]) —ñ–∑ —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—é –≤–∫–ª–∞–¥–µ–Ω—å $d_E$ —ñ —á–∞—Å–æ–≤–æ—é –∑–∞—Ç—Ä–∏–º–∫–æ—é $\tau$. –ü–æ—Ç—ñ–º –Ω–∞ –≤—ñ–¥–Ω–æ–≤–ª–µ–Ω—ñ–π —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó –º–∏ —ñ–Ω—ñ—Ü—ñ–∞–ª—ñ–∑—É—î–º–æ –ø–æ—à—É–∫ —É –ø—Ä–æ—Å—Ç–æ—Ä—ñ —Å—Ç–∞–Ω—ñ–≤ –Ω–∞–π–±–ª–∏–∂—á–æ–≥–æ –Ω–∞–π–±–ª–∏–∂—á–æ–≥–æ —Å—É—Å—ñ–¥–∞ $\vec{X}(j)$ –¥–ª—è —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó $\vec{X}(i)$:

$$
\delta_i(0) = \min_{\vec{X}(i)} \| \vec{X}(i) - \vec{X}(j) \|, \; \text{–¥–ª—è} \; \left| i-j \right| > \text{—Å–µ—Ä–µ–¥–Ω—ñ–π –ø–µ—Ä—ñ–æ–¥},
$$

–¥–µ $\| \cdot \|$ --- —Ü–µ –ï–≤–∫–ª—ñ–¥–æ–≤–∞ –Ω–æ—Ä–º–∞, $\vec{X}(j)$ --- –Ω–∞–π–±–ª–∏–∂—á–∏–π —Å—É—Å—ñ–¥–Ω—è —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—è, $\vec{X}(i)$ --- —Ä–æ–∑–≥–ª—è–¥—É–≤–∞–Ω–∞ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—è. 

–ó ([-@eq-9-1]) –º–∏ –≤–∂–µ –∑–Ω–∞—î–º–æ, —â–æ –≤—ñ–¥—Å—Ç–∞–Ω—å –º—ñ–∂ —Å—Ç–∞–Ω–∞–º–∏ $\vec{X}(i)$ —Ç–∞ $\vec{X}(j)$ –∑—Ä–æ—Å—Ç–∞—î –∑ —á–∞—Å–æ–º –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ —Å—Ç–µ–ø–µ–Ω–µ–≤–æ–≥–æ –∑–∞–∫–æ–Ω—É, –¥–µ $\lambda$ —î —Ö–æ—Ä–æ—à–∏–º –Ω–∞–±–ª–∏–∂–µ–Ω–Ω—è–º –°–ü–õ. –î–ª—è –ø–æ–¥–∞–ª—å—à–∏—Ö –æ—Ü—ñ–Ω–æ–∫ –º–∏ —Ä–æ–∑–≥–ª—è–Ω–µ–º–æ –ª–æ–≥–∞—Ä–∏—Ñ–º —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó –≤—ñ–¥—Å—Ç–∞–Ω—ñ $\ln{\delta_i(k)} \approx \lambda(k\cdot \Delta t) + \ln{c_i}$, –¥–µ $\delta_i(k)$ --- –≤—ñ–¥—Å—Ç–∞–Ω—å –º—ñ–∂ $i$-–æ—é –ø–∞—Ä–æ—é –Ω–∞–π–±–ª–∏–∂—á–∏—Ö —Å—É—Å—ñ–¥—ñ–≤, –≤–∏–∑–Ω–∞—á–µ–Ω–∏—Ö —É ([-@eq-9-6]) —á–µ—Ä–µ–∑ $k$ —á–∞—Å–æ–≤–∏—Ö –∫—Ä–æ–∫—ñ–≤, $c_i$ --- –ø–æ—á–∞—Ç–∫–æ–≤–∞ –≤—ñ–¥—Å—Ç–∞–Ω—å –º—ñ–∂ –Ω–∏–º–∏, –∞ $\Delta t$ --- —á–∞—Å–æ–≤–∏–π —ñ–Ω—Ç–µ—Ä–≤–∞–ª –º—ñ–∂ –≤–∏–º—ñ—Ä—é–≤–∞–Ω–Ω—è–º–∏ (–ø–µ—Ä—ñ–æ–¥ –¥–∏—Å–∫—Ä–µ—Ç–∏–∑–∞—Ü—ñ—ó —á–∞—Å–æ–≤–æ–≥–æ —Ä—è–¥—É).

–ü–æ–¥–∞–ª—å—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Ü—å–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—î —Ñ—É–Ω–∫—Ü—ñ—é –≤—ñ–¥ —á–∞—Å—É

$$
y(k, \Delta t) = \frac{1}{\Delta t}\frac{1}{M}\sum_{i=1}^{M}\ln{\delta_i(k)}, 
$$

–¥–µ $M=N-(d_E-1)\tau$ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—î —Ä–æ–∑–º—ñ—Ä —Ä–µ–∫–æ–Ω—Å—Ç—Ä—É–π–æ–≤–∞–Ω–æ–≥–æ —á–∞—Å–æ–≤–æ–≥–æ —Ä—è–¥—É, –∞ $\delta_i(k)$ --- –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—î $i$-—É –ª—ñ–Ω—ñ—é, –Ω–∞—Ö–∏–ª –∫–æ—Ç—Ä–æ—ó –ø—Ä–∏–±–ª–∏–∑–Ω–æ —Ä—ñ–≤–Ω–∏–π –°–ü–õ. –¢–æ–¥—ñ –ø—Ä–æ–ø–æ–Ω—É—î—Ç—å—Å—è –æ–±—á–∏—Å–ª—é–≤–∞—Ç–∏ –°–ü–õ —è–∫ –∫—É—Ç –Ω–∞—Ö–∏–ª—É –Ω–∞–π–±—ñ–ª—å—à –ª—ñ–Ω—ñ–π–Ω–æ—ó –¥—ñ–ª—è–Ω–∫–∏. –ó–Ω–∞—Ö–æ–¥–∂–µ–Ω–Ω—è —Ç–∞–∫–æ—ó –¥—ñ–ª—è–Ω–∫–∏ –≤–∏—è–≤–ª—è—î—Ç—å—Å—è –Ω–µ—Ç—Ä–∏–≤—ñ–∞–ª—å–Ω–æ—é –∑–∞–¥–∞—á–µ—é, –∞ —ñ–Ω–æ–¥—ñ –≤–∑–∞–≥–∞–ª—ñ –Ω–µ–º–æ–∂–ª–∏–≤–æ –≤–∫–∞–∑–∞—Ç–∏ —Ç–∞–∫—É –¥—ñ–ª—è–Ω–∫—É. –ù–µ–∑–≤–∞–∂–∞—é—á–∏ –Ω–∞ —Ü—é –ø—Ä–æ–±–ª–µ–º—É, –º–µ—Ç–æ–¥ –†–æ–∑–µ–Ω—à—Ç–µ–π–Ω–∞ —î –ø—Ä–æ—Å—Ç–∏–º –¥–ª—è —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—ó —Ç–∞ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è.

–í–∏–∑–Ω–∞—á–∏–º–æ —Ñ—É–Ω–∫—Ü—ñ—é `transformation()` –¥–ª—è –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ä—è–¥—É –¥–æ –ø—Ä–∏–±—É—Ç–∫–æ–≤–æ—Å—Ç–µ–π –∞–±–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–æ–≤–∞–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω—å:

```python
def transformation(signal, ret_type):

    for_rec = signal.copy()

    if ret_type == 1:       # –ó–≤–∞–∂–∞—é—á–∏ –Ω–∞ –≤–∏–¥ —Ä—è–¥—É, –≤–∏–∫–æ–Ω—É—î–º–æ
                            # –Ω–µ–æ–±—Ö—ñ–¥–Ω—ñ –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è
        pass
    elif ret_type == 2:
        for_rec = for_rec.diff()
    elif ret_type == 3:
        for_rec = for_rec.pct_change()
    elif ret_type == 4:
        for_rec = for_rec.pct_change()
        for_rec -= for_rec.mean()
        for_rec /= for_rec.std()
    elif ret_type == 5: 
        for_rec = for_rec.pct_change()
        for_rec -= for_rec.mean()
        for_rec /= for_rec.std()
        for_rec = for_rec.abs()
    elif ret_type == 6:
        for_rec -= for_rec.mean()
        for_rec /= for_rec.std()

    for_rec = for_rec.dropna().values

    return for_rec
```

–í–∏–∑–Ω–∞—á–∏–º–æ —Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è –ø–æ–±—É–¥–æ–≤–∏ –ø–∞—Ä–Ω–∏—Ö –≥—Ä–∞—Ñ—ñ–∫—ñ–≤:

```python
def plot_pair(x_values, 
              y1_values,
              y2_values,  
              y1_label, 
              y2_label,
              x_label, 
              file_name, clr="magenta"):

    fig, ax = plt.subplots()

    ax2 = ax.twinx()

    ax2.spines.right.set_position(("axes", 1.03))

    p1, = ax.plot(x_values, 
                  y1_values, 
                  "b-", label=fr"{y1_label}")
    p2, = ax2.plot(x_values,
                   y2_values, 
                   color=clr, 
                   label=y2_label)

    ax.set_xlabel(x_label)
    ax.set_ylabel(f"{y1_label}")

    ax.yaxis.label.set_color(p1.get_color())
    ax2.yaxis.label.set_color(p2.get_color())

    tkw = dict(size=2, width=1.5)

    ax.tick_params(axis='x', **tkw)
    ax.tick_params(axis='y', colors=p1.get_color(), **tkw)
    ax2.tick_params(axis='y', colors=p2.get_color(), **tkw)


    ax2.legend(handles=[p1, p2])

    plt.savefig(file_name + ".jpg")
        
    plt.show();
```

```python
signal = time_ser.copy()
ret_type = 1    # –≤–∏–¥ —Ä—è–¥—É: 
                # 1 - –≤–∏—Ö—ñ–¥–Ω–∏–π, 
                # 2 - –¥–µ—Ç—Ä–µ–Ω–¥–æ–≤–∞–Ω–∏–π (—Ä—ñ–∑–Ω–∏—Ü—è –º—ñ–∂ —Ç–µ–ø. –∑–Ω–∞—á–µ–Ω–Ω—è–º —Ç–∞ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ–º)
                # 3 - –ø—Ä–∏–±—É—Ç–∫–æ–≤–æ—Å—Ç—ñ –∑–≤–∏—á–∞–π–Ω—ñ, 
                # 4 - —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–æ–≤–∞–Ω—ñ –ø—Ä–∏–±—É—Ç–∫–æ–≤–æ—Å—Ç—ñ, 
                # 5 - –∞–±—Å–æ–ª—é—Ç–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è (–≤–æ–ª–∞—Ç–∏–ª—å–Ω–æ—Å—Ç—ñ)
                # 6 - —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–æ–≤–∞–Ω–∏–π —Ä—è–¥

time_ser_ret = transformation(signal, ret_type) 
```

–î–∞–ª—ñ –≤–∏–∑–Ω–∞—á–∏–º–æ –Ω–∞—Å—Ç—É–ø–Ω—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏:

```python
d_E = 3                      # —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å –≤–∫–ª–∞–¥–µ–Ω—å
tau = 10                     # —á–∞—Å–æ–≤—ñ –∑–∞—Ç—Ä–∏–º–∫–∞
approach_lyap = "makowski"   # –º–µ—Ç–æ–¥ –¥–ª—è —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É —Å—Ç–∞—Ä—à–æ–≥–æ –ø–æ–∫–∞–∑–Ω–∏–∫–∞
max_len = "auto"             # –≤—Å—Ç–∞–Ω–æ–≤–ª—é—î–º–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É –¥–æ–≤–∂–∏–Ω—É —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó —É 10 —Ä–∞–∑—ñ–≤ –±—ñ–ª—å—à—É –∑–∞ –∑–∞—Ç—Ä–∏–º–∫—É
sep = "auto"                 # –æ—Ü—ñ–Ω–∫–∞ —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ –ø–µ—Ä—ñ–æ–¥—É —è–∫ –≤–µ–ª–∏—á–∏–Ω—É, –æ–±–µ—Ä–Ω–µ–Ω—É –¥–æ —Å–µ—Ä–µ–¥–Ω—å–æ—ó —á–∞—Å—Ç–æ—Ç–∏ —Å–ø–µ–∫—Ç—Ä–∞ –ø–æ—Ç—É–∂–Ω–æ—Å—Ç—ñ 
```

—ñ –≤—ñ–∑—É–∞–ª—ñ–∑—É—î–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç:

```python
#| label: fig-dji-divergence 
#| fig-cap: "–î—ñ–∞–≥—Ä–∞–º–∞ —Ä–æ–∑–±—ñ–∂–Ω–æ—Å—Ç—ñ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ–π —Ä–µ–∫–æ–Ω—Å—Ç—Ä—É–π–æ–≤–∞–Ω–æ–≥–æ —Ñ–∞–∑–æ–≤–æ–≥–æ –ø—Ä–æ—Å—Ç–æ—Ä—É —ñ–Ω–¥–µ–∫—Å—É –î–æ—É –î–∂–æ–Ω—Å–∞,
#| —â–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—î —Ä–æ–∑—Ä–∞—Ö–æ–≤–∞–Ω–∏–π –°–ü–õ"

lle, _ = nk.complexity_lyapunov(signal=time_ser_ret, 
                                method=approach_lyap, 
                                dimension=d_E, 
                                delay=tau,
                                max_length=max_len,
                                separation=sep, 
                                show=True)
```

–ù–∞ [@fig-dji-divergence] –ø–æ–∫–∞–∑–∞–Ω–æ —Ç–∏–ø–æ–≤–∏–π –≥—Ä–∞—Ñ—ñ–∫ (—Å—É—Ü—ñ–ª—å–Ω–∞ –∫—Ä–∏–≤–∞) –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ —Å–µ—Ä–µ–¥–Ω—å–æ—ó —Ä–æ–∑–±—ñ–∂–Ω–æ—Å—Ç—ñ —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ–π –≤—ñ–¥ —á–∞—Å—É $\Delta t$; –ø–æ–º–∞—Ä–∞–Ω—á–µ–≤–∞ –ª—ñ–Ω—ñ—è –º–∞—î –Ω–∞—Ö–∏–ª, —â–æ –¥–æ—Ä—ñ–≤–Ω—é—î —Ç–µ–æ—Ä–µ—Ç–∏—á–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–Ω—é $\lambda_{max}$. –ö–æ—Ä–æ—Ç–∫–∞ —Å–∏–Ω—è –¥—ñ–ª—è–Ω–∫–∞ –¥–æ –ø–µ—Ä–µ—Ö–æ–¥—É —á–µ—Ä–µ–∑ —á–µ—Ä–≤–æ–Ω—É –ø—É–Ω–∫—Ç–∏—Ä–Ω—É –ª—ñ–Ω—ñ—é –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –≤–∏–ª—É—á–µ–Ω–Ω—è –Ω–∞–π–±—ñ–ª—å—à–æ–≥–æ –ø–æ–∫–∞–∑–Ω–∏–∫–∞ –õ—è–ø—É–Ω–æ–≤–∞. –Ø–∫ –º–∏ –º–æ–∂–µ–º–æ –±–∞—á–∏—Ç–∏, –∫—Ä–∏–≤–∞ –∑–º—ñ–Ω—é—î—Ç—å—Å—è –ø—Ä–∏ –±—ñ–ª—å—à–∏—Ö —á–∞—Å–æ–≤–∏—Ö –ø–µ—Ä—ñ–æ–¥–∞—Ö, –æ—Å–∫—ñ–ª—å–∫–∏ —Å–∏—Å—Ç–µ–º–∞ –æ–±–º–µ–∂–µ–Ω–∞ —É —Ñ–∞–∑–æ–≤–æ–º—É –ø—Ä–æ—Å—Ç–æ—Ä—ñ —ñ —Å–µ—Ä–µ–¥–Ω—è –¥–∏–≤–µ—Ä–≥–µ–Ω—Ü—ñ—è –Ω–µ –º–æ–∂–µ –ø–µ—Ä–µ–≤–∏—â—É–≤–∞—Ç–∏ "–¥–æ–≤–∂–∏–Ω—É" –∞—Ç—Ä–∞–∫—Ç–æ—Ä–∞. –û—Ç—Ä–∏–º–∞–Ω–∏–π –ø–æ–∫–∞–∑–Ω–∏–∫ –õ—è–ø—É–Ω–æ–≤–∞ –≤–∫–∞–∑—É—î –Ω–∞ —Ç–µ, —â–æ —ñ–Ω–¥–µ–∫—Å –î–æ—É –î–∂–æ–Ω—Å–∞ –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –Ω–∞ –º–µ–∂—ñ –º—ñ–∂ —Ö–∞–æ—Å–æ–º —Ç–∞ —Å—Ç–∞–±—ñ–ª—å–Ω—ñ—Å—Ç—å, —Ç–æ–±—Ç–æ —ñ–Ω–¥–µ–∫—Å –¥–∏–≤–µ—Ä–≥–µ–Ω—Ü—ñ—è –¥–∏–Ω–∞–º—ñ–∫–∏ —Ä—è–¥—É –≤—Ä—ñ–≤–Ω–æ–≤–∞–∂—É—î—Ç—å—Å—è –∫–æ–Ω–≤–µ—Ä–≥–µ–Ω—Ü—ñ—î—é. 

–Ø–∫ –º–∏ –≤–∂–µ –º–∞–ª–∏ –∑–º–æ–≥—É –ø–µ—Ä–µ–∫–æ–Ω–∞—Ç–∏—Å—å, —Å–∫–ª–∞–¥–Ω—ñ —Å–∏—Å—Ç–µ–º–∏ –º—ñ–Ω–ª–∏–≤—ñ, —â–æ –≥–æ–≤–æ—Ä–∏—Ç—å –ø—Ä–æ —Ç–µ, —â–æ —Å–∏—Å—Ç–µ–º–∞ –∑ –ø–ª–∏–Ω–æ–º —á–∞—Å—É –º–æ–∂–µ –ø—Ä–æ—è–≤–ª—è—Ç–∏ —è–∫ –∫–æ–Ω–≤–µ—Ä–≥–µ–Ω—Ü—ñ—é, —Ç–∞–∫ —ñ –¥–∏–≤–µ—Ä–≥–µ–Ω—Ü—ñ—é, —Ç–∞–∫ —ñ –ø–æ–≤–Ω—É –Ω–µ–∑–º—ñ–Ω–Ω—ñ—Å—Ç—å —ñ–∑ –ø–ª–∏–Ω–æ–º —á–∞—Å—É. 

–î–∞–ª—ñ —Ä–æ–∑–≥–ª—è–Ω–µ–º–æ –¥–∏–Ω–∞–º—ñ–∫—É –¥–æ—Å–ª—ñ–¥–∂—É–≤–∞–Ω–æ—ó —Å–∏—Å—Ç–µ–º–∏ –∑ –ø–ª–∏–Ω–æ–º —á–∞—Å—É –≤ —Ä–∞–º–∫–∞—Ö –ø—Ä–æ—Ü–µ–¥—É—Ä–∏ **–∫–æ–≤–∑–Ω–æ–≥–æ –≤—ñ–∫–Ω–∞**. –í–∏–∑–Ω–∞—á–∏–º–æ –Ω–∞—Å—Ç—É–ø–Ω—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏:

```python
window = 500            # —à–∏—Ä–∏–Ω–∞ –≤—ñ–∫–Ω–∞
tstep = 1               # —á–∞—Å–æ–≤–∏–π –∫—Ä–æ–∫ –≤—ñ–∫–Ω–∞ 
length = len(time_ser)  # –¥–æ–≤–∂–∏–Ω–∞ —Å–∞–º–æ–≥–æ —Ä—è–¥—É
ret_type = 1            # –≤–∏–¥ —Ä—è–¥—É: 
                        # 1 - –≤–∏—Ö—ñ–¥–Ω–∏–π, 
                        # 2 - –¥–µ—Ç—Ä–µ–Ω–¥–æ–≤–∞–Ω–∏–π (—Ä—ñ–∑–Ω–∏—Ü—è –º—ñ–∂ —Ç–µ–ø. –∑–Ω–∞—á–µ–Ω–Ω—è–º —Ç–∞ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ–º)
                        # 3 - –ø—Ä–∏–±—É—Ç–∫–æ–≤–æ—Å—Ç—ñ –∑–≤–∏—á–∞–π–Ω—ñ, 
                        # 4 - —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–æ–≤–∞–Ω—ñ –ø—Ä–∏–±—É—Ç–∫–æ–≤–æ—Å—Ç—ñ, 
                        # 5 - –∞–±—Å–æ–ª—é—Ç–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è (–≤–æ–ª–∞—Ç–∏–ª—å–Ω–æ—Å—Ç—ñ)
                        # 6 - —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–æ–≤–∞–Ω–∏–π —Ä—è–¥

d_E = 3                      # —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å –≤–∫–ª–∞–¥–µ–Ω—å
tau = 1                      # —á–∞—Å–æ–≤—ñ –∑–∞—Ç—Ä–∏–º–∫–∞
approach_lyap = "makowski"   # –º–µ—Ç–æ–¥ –¥–ª—è —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É —Å—Ç–∞—Ä—à–æ–≥–æ –ø–æ–∫–∞–∑–Ω–∏–∫–∞: rosenstein1993, makowski 
max_len = "auto"             # –≤—Å—Ç–∞–Ω–æ–≤–ª—é—î–º–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É –¥–æ–≤–∂–∏–Ω—É —Ç—Ä–∞—î–∫—Ç–æ—Ä—ñ—ó —É 10 —Ä–∞–∑—ñ–≤ –±—ñ–ª—å—à—É –∑–∞ –∑–∞—Ç—Ä–∏–º–∫—É: auto
sep = "auto"                 # –æ—Ü—ñ–Ω–∫–∞ —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ –ø–µ—Ä—ñ–æ–¥—É —è–∫ –≤–µ–ª–∏—á–∏–Ω—É, –æ–±–µ—Ä–Ω–µ–Ω—É –¥–æ —Å–µ—Ä–µ–¥–Ω—å–æ—ó —á–∞—Å—Ç–æ—Ç–∏ —Å–ø–µ–∫—Ç—Ä–∞ –ø–æ—Ç—É–∂–Ω–æ—Å—Ç—ñ

LLE = []                # –º–∞—Å–∏–≤ –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è –°–ü–õ
```

–¢–µ–ø–µ—Ä –º–æ–∂–Ω–∞ –ø—Ä–∏—Å—Ç—É–ø–∞—Ç–∏ –¥–æ –≤—ñ–∫–æ–Ω–Ω–æ—ó –ø—Ä–æ—Ü–µ–¥—É—Ä–∏:

```python
for i in tqdm(range(0,length-window,tstep)):  # —Ñ—Ä–∞–≥–º–µ–Ω—Ç–∏ –¥–æ–≤–∂–∏–Ω–æ—é window  
                                              # –∑ –∫—Ä–æ–∫–æ–º tstep

    fragm = time_ser.iloc[i:i+window].copy()  # –≤—ñ–¥–±–∏—Ä–∞—î–º–æ —Ñ—Ä–∞–≥–º–µ–Ω—Ç

    fragm = transformation(fragm, ret_type)   # –≤–∏–∫–æ–Ω—É—î–º–æ –ø—Ä–æ—Ü–µ–¥—É—Ä—É 
                                              # —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü—ñ—ó —Ä—è–¥—É
    
    lle, _ = nk.complexity_lyapunov(signal=fragm, 
                                method=approach_lyap, 
                                dimension=d_E, 
                                delay=tau,
                                max_length=max_len,
                                separation=sep, 
                                show=False)
    
    LLE.append(lle)
```

–ó–±–µ—Ä—ñ–≥–∞—î–º–æ –æ—Ç—Ä–∏–º–∞–Ω—ñ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ –≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–º—É —Ñ–∞–π–ª—ñ:

```python
name = f"LLE_name={symbol}_window={window}_step={tstep}_rettype={ret_type}_\
    d_E={d_E}_tau={tau}_approach={approach_lyap}_max_len={max_len}_separation={sep}.txt"

np.savetxt(name, LLE)
```

–í–∏–∑–Ω–∞—á–∞—î–º–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏ –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è —Ä–∏—Å—É–Ω–∫—ñ–≤:

```python
# –ø–æ–∑–Ω–∞—á–µ–Ω–Ω—è –ø–æ–∫–∞–∑–Ω–∏–∫–∞ –õ—è–ø—É–Ω–æ–≤–∞ –≤ –ª–µ–≥–µ–Ω–¥—ñ —Ä–∏—Å—É–Ω–∫—É 
label_lyap = r'$\lambda_{max}$'  

# –Ω–∞–∑–≤–∞ —Ä–∏—Å—É–Ω–∫—É
file_name = f"LLE_name={symbol}_window={window}_step={tstep}_rettype={ret_type}_\
    d_E={d_E}_tau={tau}_approach={approach_lyap}_max_len={max_len}_separation={sep}"

# –∫–æ–ª—ñ—Ä –ø–æ–∫–∞–∑–Ω–∏–∫–∞
color = 'red'  
```

—Ç–∞ –≤–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç:

```python
#| label: fig-dji-wind-lle  
#| fig-cap: "–î–∏–Ω–∞–º—ñ–∫–∞ —ñ–Ω–¥–µ–∫—Å—É –î–æ—É –î–∂–æ–Ω—Å–∞ —Ç–∞ 
#| —Å—Ç–∞—Ä—à–æ–≥–æ –ø–æ–∫–∞–∑–Ω–∏–∫–∞ –õ—è–ø—É–Ω–æ–≤–∞"

plot_pair(time_ser.index[window:length:tstep],
          time_ser.values[window:length:tstep],
          LLE, 
          ylabel, 
          label_lyap,
          xlabel,
          file_name,
          color)
```

–Ø–∫ –º–∏ –º–æ–∂–µ–º–æ –±–∞—á–∏—Ç–∏ –Ω–∞ [@fig-dji-wind-lle], –°–ü–õ –ø–æ—á–∏–Ω–∞—î —Å–ø–∞–¥–∞—Ç–∏ –≤ –∫—Ä–∏–∑–æ–≤—ñ —Ç–∞ –ø–µ—Ä–µ–¥–∫—Ä–∏–∑–æ–≤—ñ —Å—Ç–∞–Ω–∏, —â–æ –≤–∫–∞–∑—É—î –Ω–∞ –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è –∫–æ—Ä–µ–ª—å–æ–≤–∞–Ω–Ω–æ—Å—Ç—ñ –¥–æ—Å–ª—ñ–¥–∂—É–≤–∞–Ω–æ—ó –¥–∏–Ω–∞–º—ñ–∫–∏. –£ –º–æ–º–µ–Ω—Ç –∫—Ä–∏–∑–∏ –°–ü–õ –ø–æ—á–∏–Ω–∞—î –∑—Ä–æ—Å—Ç–∞—Ç–∏, —â–æ –≤–∫–∞–∑—É—î –Ω–∞ –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è –¥–∏–≤–µ—Ä–≥–µ–Ω—Ü—ñ—ó –≤ –∫—Ä–∏–∑–æ–≤—ñ –ø–µ—Ä—ñ–æ–¥–∏. 

#### –û–±—á–∏—Å–ª–µ–Ω–Ω—è –ø–æ–∫–∞–∑–Ω–∏–∫—ñ–≤ –õ—è–ø—É–Ω–æ–≤–∞ –Ω–∞ –æ—Å–Ω–æ–≤—ñ –º–µ—Ç–æ–¥—É –ï–∫–º–∞–Ω–∞

–†–æ–∑–≥–ª—è–Ω–µ–º–æ –¥–∏–Ω–∞–º—ñ–∫—É –ø–æ–∫–∞–∑–Ω–∏–∫—ñ–≤ –õ—è–ø—É–Ω–æ–≤–∞ –≤ –º–µ–∂–∞—Ö –ø—Ä–æ—Ü–µ–¥—É—Ä–∏ —Ä—É—Ö–æ–º–æ–≥–æ –≤—ñ–∫–Ω–∞. –í–∏–∑–Ω–∞—á–∏–º–æ –Ω–∞—Å—Ç—É–ø–Ω—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏:

```python
window = 500            # —à–∏—Ä–∏–Ω–∞ –≤—ñ–∫–Ω–∞
tstep = 1               # —á–∞—Å–æ–≤–∏–π –∫—Ä–æ–∫ –≤—ñ–∫–Ω–∞ 
length = len(time_ser)  # –¥–æ–≤–∂–∏–Ω–∞ —Å–∞–º–æ–≥–æ —Ä—è–¥—É
ret_type = 1            # –≤–∏–¥ —Ä—è–¥—É: 
                        # 1 - –≤–∏—Ö—ñ–¥–Ω–∏–π, 
                        # 2 - –¥–µ—Ç—Ä–µ–Ω–¥–æ–≤–∞–Ω–∏–π (—Ä—ñ–∑–Ω–∏—Ü—è –º—ñ–∂ —Ç–µ–ø. –∑–Ω–∞—á–µ–Ω–Ω—è–º —Ç–∞ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ–º)
                        # 3 - –ø—Ä–∏–±—É—Ç–∫–æ–≤–æ—Å—Ç—ñ –∑–≤–∏—á–∞–π–Ω—ñ, 
                        # 4 - —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–æ–≤–∞–Ω—ñ –ø—Ä–∏–±—É—Ç–∫–æ–≤–æ—Å—Ç—ñ, 
                        # 5 - –∞–±—Å–æ–ª—é—Ç–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è (–≤–æ–ª–∞—Ç–∏–ª—å–Ω–æ—Å—Ç—ñ)
                        # 6 - —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–æ–≤–∞–Ω–∏–π —Ä—è–¥

d_E = 4                         # —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å –≤–∫–ª–∞–¥–µ–Ω—å –≤–∏—Ö—ñ–¥–Ω–æ–≥–æ –ø—Ä–æ—Å—Ç–æ—Ä—É (–∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ–∫–∞–∑–Ω–∏–∫—ñ–≤)
d_M = 3                         # —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å –≤–∫–ª–∞–¥–µ–Ω—å –ø—ñ–¥–ø—Ä–æ—Å—Ç–æ—Ä—É 

approach_lyap = "eckmann1986"   # –º–µ—Ç–æ–¥ –¥–ª—è —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É —Å—Ç–∞—Ä—à–æ–≥–æ –ø–æ–∫–∞–∑–Ω–∏–∫–∞
sep = "auto"                    # –æ—Ü—ñ–Ω–∫–∞ —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ –ø–µ—Ä—ñ–æ–¥—É —è–∫ –≤–µ–ª–∏—á–∏–Ω—É, –æ–±–µ—Ä–Ω–µ–Ω—É –¥–æ —Å–µ—Ä–µ–¥–Ω—å–æ—ó —á–∞—Å—Ç–æ—Ç–∏ —Å–ø–µ–∫—Ç—Ä–∞ –ø–æ—Ç—É–∂–Ω–æ—Å—Ç—ñ
min_neighb = "default"          # min(2 * matrix_dim, matrix_dim + 4)        

LE = []                         # –º–∞—Å–∏–≤ –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è –ü–õ  
```

–¢–µ–ø–µ—Ä –ø–µ—Ä–µ—Ö–æ–¥–∏–º–æ –¥–æ —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—ñ–≤:

```python
for i in tqdm(range(0,length-window,tstep)):  # —Ñ—Ä–∞–≥–º–µ–Ω—Ç–∏ –¥–æ–≤–∂–∏–Ω–æ—é window  
                                              # –∑ –∫—Ä–æ–∫–æ–º tstep

    fragm = time_ser.iloc[i:i+window].copy()  # –≤—ñ–¥–±–∏—Ä–∞—î–º–æ —Ñ—Ä–∞–≥–º–µ–Ω—Ç

    fragm = transformation(fragm, ret_type)   # –≤–∏–∫–æ–Ω—É—î–º–æ –ø—Ä–æ—Ü–µ–¥—É—Ä—É 
                                              # —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü—ñ—ó —Ä—è–¥—É
    
    le, _ = nk.complexity_lyapunov(signal=fragm, 
                                method=approach_lyap, 
                                dimension=d_E,
                                matrix_dim=d_M, 
                                min_neighbors=min_neighb,
                                separation=sep, 
                                show=False)

    LE.append(le)
```

–ó–±–µ—Ä—ñ–≥–∞—î–º–æ –æ—Ç—Ä–∏–º–∞–Ω—ñ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ –≤ —Ç–µ–∫—Å—Ç–æ–≤–∏—Ö —Ñ–∞–π–ª–∞—Ö:

```python
LE = np.array(LE)

for i in range(d_E):
    np.savetxt(f"LE number={i+1}_name={symbol}_window={window}_step={tstep}_rettype={ret_type}_\
    d_E={d_E}_d_M={d_M}_approach={approach_lyap}_min_neighbors={min_neighb}_separation={sep}.txt", LE[i])
```

```python
#| label: fig-dji-wind-spec-le  
#| fig-cap: "–î–∏–Ω–∞–º—ñ–∫–∞ —ñ–Ω–¥–µ–∫—Å—É –î–æ—É –î–∂–æ–Ω—Å–∞ —Ç–∞ —Å–ø–µ–∫—Ç—Ä–∞ –ø–æ–∫–∞–∑–Ω–∏–∫—ñ–≤ –õ—è–ø—É–Ω–æ–≤–∞"


fig, ax = plt.subplots(LE.shape[1]+1, 1, figsize=(10, 10), sharex=True)

ax[0].plot(time_ser.index[window:length:tstep], time_ser.values[window:length:tstep], label=symbol)
ax[0].set_ylabel(symbol)
ax[0].legend()

for i in range(1, LE.shape[1]+1):
    ax[i].plot(time_ser.index[window:length:tstep], LE[:, i-1], color='red', label=fr'$\lambda_{i}$')
    ax[i].set_ylabel(fr"$\lambda_{i}$")
    ax[i].legend()

ax[-1].set_xlabel(xlabel)
fig.subplots_adjust(hspace=0)

plt.savefig(f"LE name={symbol}_window={window}_step={tstep}_rettype={ret_type}_\
    #d_E={d_E}_d_M={d_M}_approach={approach_lyap}_min_neighbors={min_neighb}_separation={sep}.jpg")
plt.show();
```
